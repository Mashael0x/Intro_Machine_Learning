## **2. Encoding categorical features**

```
data.info()
```

```
type_encoded = pd.get_dummies(data['type'])
type_encoded
```

```
location_encoded = pd.get_dummies(data['location'])
location_encoded
```

```
data.title
```

```
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
```

```
title_transformed = cv.fit_transform(data.title)
```

```
title_transformed = pd.DataFrame(title_transformed.toarray(), columns= cv.get_feature_names())
```

```
title_transformed.head()
```

```
title_transformed.shape
```

```
data = pd.concat([data,type_encoded,location_encoded,title_transformed], axis= 1)
data = data.drop(columns = ['ds','search_filter','url','title','type','location'], axis = 1)
```
